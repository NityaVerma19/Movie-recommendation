### *DATA 1*

*Merging data1 with movie titles*

```python
data1_titles = data1.merge(movie_title, on = 'MovieID', how = 'inner')
```

```python
data1_titles.head()
```

```python
data1_titles.nunique()
```

*Merging data1 titles with `netflix titles`*

```python
data1_merged = net.merge(data1_titles, how = 'inner', on = 'title')
```

```python
data1_merged.shape
```

```python
data1_merged.head()
```

```python
data1_merged.nunique()
```

*Finding the `Count of ratings`*

```python
rating_count_1 = data1_merged.groupby('CustomerID')['Rating'].count().sort_values(ascending=False)

rating_count_1.head()
```

*Filtering data by removing `outliers`*

```python
q1_1 = rating_count_1.quantile(0.25)
q3_1 = rating_count_1.quantile(0.75)

iqr_1 = q3_1 - q1_1
upper_bound_1 = q3_1 + iqr_1 * 1.5

filtered_df_1 = data1_merged[~data1_merged['CustomerID'].isin([x for x in rating_count_1.index if rating_count_1[x] > upper_bound_1])]

filtered_df_1.shape
```

```python
filtered_df_1['Rating'].describe()
```
