```python
data_num_ratings['num_ratings'] = data_num_ratings.groupby('title')['title'].transform('count')
data_num_ratings = data_num_ratings.sort_values(by = ["Rating",'num_ratings'], ascending = [False,False])
data_num_ratings = data_num_ratings.drop_duplicates(subset =['title', 'MovieID'],keep = "first")
data_num_ratings.drop(columns=['CustomerID', 'MovieID'], inplace=True)
data_num_ratings
```

# *COLLABORATIVE FILTERING*

### **STEPS**

1.   Make user-movie matrix
2.   Find Similarity between users
3.   Make a movie-movie dataframe
4.   Formulate the recommendation function
5.   Recommend



**`STEP 1`**

```python
fd.head()
```

```python
user_movie_interact = fd.pivot_table(index = 'title', columns = 'CustomerID', values = 'Rating')
user_movie_interact.fillna(0, inplace = True)
user_movie_interact
```

*Reducing the size of values*

```python
user_movie_interact = user_movie_interact.astype('int8')
```

**`STEP 2- FINDING SIMILARITY`**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(user_movie_interact)
similarity_score
```

```python
similarity_score.shape
```

**`STEP 3- MAKE A MOVIE-MOVIE DF`**

```python
similar_movie_df = pd.DataFrame(similarity_score, index = user_movie_interact.index, columns = user_movie_interact.index)
similar_movie_df
```

**`STEP 4- RECOMMENDATION FUNCTION`**

```python
def recommend(movie_title, rating):
    if movie_title in similar_movie_df.index:
        score = similar_movie_df[movie_title]*(rating-2.5)
        score = score.sort_values(ascending = False)[1:6]    #recommend top 5 movies

        return score

    else:
      top_movies = data_num_ratings[:6]
      print(f'"{movie_title}" not found in the database')
      print("Watch these instead:")
      return top_movies[['title', 'num_ratings']]
```

Collaborative filtering, as a recommendation technique, doesn't inherently require recommending items of the same genre. Collaborative filtering algorithms recommend items based on user behavior and preferences, typically by finding similarities between users or items.

The recommendation process in collaborative filtering works by identifying users who have similar preferences and tastes to a target user and then recommending items that those similar users have liked or interacted with in the past. Similarly, it can also identify items that are similar to those the target user has interacted with before.

While genre can be one factor used to determine similarity between items or users, it's not the only factor. Collaborative filtering algorithms can consider various aspects of items and user behavior, such as ratings, purchase history, viewing history, and so on.
